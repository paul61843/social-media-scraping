```python
# pip3 install seleniumbase
from seleniumbase import Driver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By

import time
import random
```

```python
# initialize the driver in GUI mode with UC enabled
driver = Driver(uc=True, headless=False)

# set the target URL
url = "https://www.dcard.tw/f/stock?tab=latest"
```

```python
# set the stop date EX 8/25
stop_date = "8/25"

# open URL with a 6-second reconnect time to bypass the initial JS challenge
driver.uc_open_with_reconnect(url, reconnect_time=6)

while True:
    driver.execute_script("window.scrollTo({ top: document.body.scrollHeight, behavior: 'smooth' });")
    time.sleep(random.uniform(5, 7))  # 等待 10 秒

    # check date time is "2024-06-20"
    articles = driver.find_elements(By.CSS_SELECTOR, "time")
    last_article = articles[-1]
    date_time = last_article.get_attribute("innerText")
    if date_time == stop_date:
        break
```

```python
# scraping html articles to json
articles = driver.find_elements(By.CSS_SELECTOR, "div[role='main'] section > div")

data = []
for article in articles:
    try:
        title = article.find_element(By.CSS_SELECTOR, "h2").get_attribute("innerText")
    except:
        title = ""
    try:
        outline = article.find_element(By.CSS_SELECTOR, "p").get_attribute("innerText")
    except:
        outline = ""
    try:
        date_time = article.find_element(By.CSS_SELECTOR, "time").get_attribute("innerText")
    except:
        date_time = ""
    try:
        link = article.find_element(By.CSS_SELECTOR, "a:has(time)").get_attribute("href")
    except:
        link = ""
    data.append({"title": title, "outline": outline, "date_time": date_time, "link": link})

# data to json file
import json
with open("dcard_stock.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

# close the driver
# driver.quit()
```
